```python
# CASE.01 - urllib.request.urlopen 을 사용할 경우
from urllib.request import urlopen

html_urllib = urlopen(url_target)
response_urllib = BeautifulSoup(html_urllib, 'html.parser')
```

```python
# 정확히 똑같지는 않은데, 내용은 같다
response_requests is response_urllib
```

```python
def find_tags(tags, response, detailed=False):
    finders = response.find_all(*tags)
    print(f"find('{tags}') : {len(finders)}")

    for i, find in enumerate(finders, 1):
        text_string = find.text.replace('\n', '')
        if text_string.find('카카오'):
            print('-----------------'*5)
            if detailed:
                print(f"{i:>02}. {text_string} ... {find}")
            else:
                print(f"   {i:>02}. {text_string}")
```

```python
tags = ["a",]
tags = ["a", {"class":"link_favorsch"}] # class="link_favorsch
find_tags(tags, response_urllib)
```

```python
tags = ["a",{'class' : 'link_txt'}]
find_tags(tags, response_urllib, detailed=0)
```

```python
tags = ["a", { 'class' : 'link_txt'}]   # name="a", attrs={ 'class' : 'link_txt'}
find_tags(tags, response_urllib)
```
